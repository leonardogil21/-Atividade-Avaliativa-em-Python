Questão 1 — Classificação de
Temperaturas (Condicionais + Loop)

Dada a lista de temperaturas registradas durante a semana:
temperaturas = [18, 22, 30, 35, 28, 15, 40]
Classifique cada dia como:
• Abaixo de 20 → "Frio"
• Entre 20 e 30 → "Agradável"
• Acima de 30 → "Quente"
Tarefas:
1. Criar uma lista com as classificações
2. Contar quantos dias houve em cada categoria


```python
temperaturas = [18, 22, 30, 35, 28, 15, 40]

classificacoes = []
for temp in temperaturas:
    if temp < 20:
        classificacoes.append("Frio")
    elif 20 <= temp <= 30:
        classificacoes.append("Agradável")
    else: # temp > 30
        classificacoes.append("Quente")

print(f"Classificações: {classificacoes}")

# Contar quantos dias houve em cada categoria
contagem_categorias = {"Frio": 0, "Agradável": 0, "Quente": 0}
for classificacao in classificacoes:
    contagem_categorias[classificacao] += 1

print(f"Contagem por categoria: {contagem_categorias}")
```

Questão 2 — Sistema de Avaliação de
Alunos

Dada a lista de notas finais:
notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]
Classifique:
• < 5 → "Reprovado"
• 5 a 7 → "Recuperação"
• 7 → "Aprovado"
Tarefas:
1. Gerar lista de classificações
2. Calcular:
o Quantidade de aprovados
o Percentual de reprovados


```python
notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]

classificacoes_notas = []
for nota in notas:
    if nota < 5:
        classificacoes_notas.append("Reprovado")
    elif 5 <= nota < 7:
        classificacoes_notas.append("Recuperação")
    else: # nota >= 7
        classificacoes_notas.append("Aprovado")

print(f"Classificações das notas: {classificacoes_notas}")

# Calcular quantidade de aprovados e percentual de reprovados
quantidade_aprovados = classificacoes_notas.count("Aprovado")
quantidade_reprovados = classificacoes_notas.count("Reprovado")
total_alunos = len(notas)

percentual_reprovados = (quantidade_reprovados / total_alunos) * 100 if total_alunos > 0 else 0

print(f"Quantidade de aprovados: {quantidade_aprovados}")
print(f"Percentual de reprovados: {percentual_reprovados:.2f}%")
```

Questão 3 — Monitoramento de
Consumo de Energia

Consumo diário (kWh):
consumo = [120, 95, 200, 180, 75, 220, 160]
Regras:
• < 100 → "Baixo"
• 100–180 → "Moderado"
• 180 → "Alto"
Tarefas:
1. Filtrar dias com consumo alto
2. Calcular:
o Total de consumo da semana
o Média semanal
3. Exibir alerta se houver mais de 2 dias com consumo alto

```python
consumo = [120, 95, 200, 180, 75, 220, 160]

# 1. Filtrar dias com consumo alto
dias_alto_consumo = []
for c in consumo:
    if c >= 180: # Interpretando '180 -> "Alto"' como '>=' 180
        dias_alto_consumo.append(c)

print(f"Dias com consumo alto: {dias_alto_consumo}")

# 2. Calcular total e média semanal
total_consumo_semana = sum(consumo)
media_semanal = total_consumo_semana / len(consumo)

print(f"Total de consumo da semana: {total_consumo_semana} kWh")
print(f"Média semanal de consumo: {media_semanal:.2f} kWh")

# 3. Exibir alerta se houver mais de 2 dias com consumo alto
if len(dias_alto_consumo) > 2:
    print("ALERTA: Mais de 2 dias com consumo alto nesta semana!")
```

Questão 4 — Simulação de Carrinho de
 Compras
Lista de preços:
precos = [50, 120, 30, 200, 80, 15]
 Regras de desconto:
• Item < 50 → 5% desconto
• 50–150 → 10% desconto
• 150 → 15% desconto
Tarefas:
1. Criar lista com preços finais
2. Calcular:
o Total da compra com desconto
o Economia total obtida

```python
precos = [50, 120, 30, 200, 80, 15]

precos_finais = []
economia_total = 0

for preco in precos:
    desconto_percentual = 0
    if preco < 50:
        desconto_percentual = 0.05  # 5%
    elif 50 <= preco <= 150:
        desconto_percentual = 0.10  # 10%
    else:  # preco > 150 (interpretando '150 → 15% desconto' como > 150)
        desconto_percentual = 0.15  # 15%

    valor_desconto = preco * desconto_percentual
    preco_final = preco - valor_desconto

    precos_finais.append(preco_final)
    economia_total += valor_desconto

# Corrigido: formatar cada item da lista individualmente
print(f"Lista com preços finais: {[f'{p:.2f}' for p in precos_finais]}")

total_compra_com_desconto = sum(precos_finais)
print(f"Total da compra com desconto: {total_compra_com_desconto:.2f}")
print(f"Economia total obtida: {economia_total:.2f}")
```
